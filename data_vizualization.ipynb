Analisar tendência de imigração do Brasil em determinado período.

```python
#selecionando apenas a parte do Brasil
brasil = df.loc['Brasil', anos]
```

```python
#convertendo para um dataframe, criando um data frame criando um dicionário
brasil_dict = {'ano' : brasil.index.tolist(), 'imigrantes' : brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
#comparação de tendências de múltiplos países

fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3) #Espaçamento horizonta e vertical
fig.suptitle('Imigração dos maiores países da América do Sul para o Canadá de 1980 a 2013')

axs[0, 0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #adicionando a frequência de 5 anos

for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de imigrantes')

#colocando os eixos dos gráficos com os mesmos valores

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)


plt.show()
```

```python
# executar primeiro o no bloco abaixo para a função funcionar america_sul_sorted= america_sul.sort_values(by='Total', ascending=True)

cores = []

for pais in america_sul_sorted.index: #deixando o gráfico com o Brasil destacado
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')


fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores) #'barh' para criar grafico na horizontal
ax.set_title('América do Sul: Brasil foi  quarto país com mais imigrantes\n para o Canadá no período de 1980 a 2013', loc='left', fontsize='18') #titulo e tamanho da fonte
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize='12')
ax.yaxis.set_tick_params(labelsize='12')

#adicionando valores para as barras

for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize='10', ha='left', va='center')



ax.set_frame_on(False) #removendo o frame em volta
ax.get_xaxis().set_visible(False) #removendo o eixo X
ax.tick_params(axis='both', which='both', length=0) #removendo os ticks

fig.savefig('america_imigracao_canada', transparent=False, dpi=300, bbox_inches='tight')



plt.show()
```

```python
import seaborn as sns
```

```python
#adicionando tema

sns.set_theme()
```

```python
top_10 = df.sort_values('Total', ascending=False).head(10) #selecionando os 10 primeiro paises
```

```python
top_10
```

```python
#plotando grafico de barras

sns.barplot(data=top_10, x=top_10.index, y='Total')
```

```python
#colocando o gráfico em horizontal

sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')
```

```python
#adicionando o título

ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set(title='Paises com maio imigração para o Canadá\n1980 a 2013', xlabel='Número de imigrantes', ylabel='')

plt.show()
```

```python
#adicionando personalizações ao gráfico

fig, ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set_title('Paises com maio imigração para o Canadá\n1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
plt.show()
```

```python
#Alterando cores com paleta

def gerar_grafico_paleta(pallete):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=pallete)
  ax.set_title('Paises com maio imigração para o Canadá\n1980 a 2013', loc='left', fontsize=16)
  ax.set_xlabel('Número de imigrantes', fontsize=14)

  plt.show()
```

```python
#Conhecendo as paletas

gerar_grafico_paleta('Blues')
```

```python
#alterando o esquema de cores para a forma reversa

gerar_grafico_paleta('Blues_r')
```

```python
#usando paleta divergente

gerar_grafico_paleta('rocket')
```

```python
#outra paleta divergente

gerar_grafico_paleta('coolwarm')
```

```python
#utilizando uma paleta categórica uma para cada país

gerar_grafico_paleta('tab10')
```

```python
#Alterando temas

sns.set_theme(style='dark')
gerar_grafico_paleta('tab10')
```

```python

sns.set_theme(style='whitegrid')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks') #coloca um tick acima dos números no eixo x
gerar_grafico_paleta('tab10')
```

```python
#criando df da Argentina

argentina = df.loc['Argentina', anos]
```

```python
import plotly.graph_objs as go

# Criando uma figura
fig = go.Figure()

# Adicionando a linha com os dados do Brasil
fig.add_trace(
    go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]], mode='lines', name='Imigrantes do Brasil', line=dict(width=4))
)

# Adicionando a linha com os dados da Argentina
fig.add_trace(
    go.Scatter(x=[dados_argentina['ano'].iloc[0]], y=[dados_argentina['imigrantes'].iloc[0]], mode='lines', name='Imigrantes da Argentina', line=dict(width=4))
)

# Definindo as configurações de layout
fig.update_layout(
    title=dict(
        text='<b>Imigração do Brasil e da Argentina para o Canadá no período de 1980 a 2013',
        x=0.1,

        font=dict(size=18)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1200, # largura da figura em pixels
    height=600 # altura da figura em pixels
)

# Definindo as configurações de animação
frames = []
for i in range(len(dados_brasil)):
    frame_data = [
        go.Scatter(x=dados_brasil['ano'].iloc[:i+1], y=dados_brasil['imigrantes'].iloc[:i+1]),
        go.Scatter(x=dados_argentina['ano'].iloc[:i+1], y=dados_argentina['imigrantes'].iloc[:i+1])
    ]
    frame = go.Frame(data=frame_data)
    frames.append(frame)
fig.frames = frames

# Mostrando a figura
fig.show()
```
