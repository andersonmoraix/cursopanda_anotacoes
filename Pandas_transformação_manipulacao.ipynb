Transformação e manipulação de dados


```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

```python
#Normalizando os dados que estão aninhados
dados = pd.json_normalize(dados['info_moveis'])
dados
```

# Do que se trata cada dado:

avaliacao_geral: refere-se à média de notas dadas para a avaliação da hospedagem no imóvel.

experiencia_local: descreve as experiências oferecidas durante a hospedagem no imóvel.

max_hospedes: informa a quantidade máxima de hóspedes que o local permite.

descricao_local: descreve o imóvel.

descricao_vizinhanca: descreve a vizinhança ao redor do imóvel.

quantidade_banheiros: informa a quantidade de banheiros disponíveis.

quantidade_quartos: informa a quantidade de quartos disponíveis.

quantidade_camas: informa a quantidade de camas disponíveis.

modelo_cama: informa o modelo de cama oferecido.

comodidades: informa as comodidades oferecidas pelo imóvel.

taxa_deposito: informa a taxa de depósito mínima para segurança de hospedagem.

taxa_limpeza: informa a taxa cobrada para o serviço de limpeza.

preco: refere-se ao preço base a ser cobrado pela diária no imóvel.

```python
#Retirando listas, desagrupando dados
#coletando o nome das colunas, lista de todas as colunas que há no arquivo

colunas = list(dados.columns) #nome de todas as colunas do dataframe
colunas
```

```python
#desagrupando com o método 'explode'

dados = dados.explode(colunas[3:])#a partir da terceira coluna
dados
```

```python
#resetando os indices para reordenar os dados

dados.reset_index(inplace=True, drop=True)
dados.head()
```

```python
#verificando os tipos de dados, observando a coluna Dtype(object são varios tipos de dados diferentes)

dados.info()
```

```python
#Convertendo dados numéricos com o metodo astype tratando dados numericos convertendo para tipo inteiro
import numpy as np
```

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
#convertendo para o tipo float na coluna avaliacao geral

dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

```python
#verificando se o 'astype' funciona com '$' ou ','

dados['taxa_deposito'] = dados['taxa_deposito'].apply(lambda x: x.replace('$','').replace(',', '').strip())
dados['taxa_limpeza'] = dados['taxa_limpeza'].apply(lambda x: x.replace('$','').replace(',', '').strip())

dados['taxa_deposito'] = dados['taxa_deposito'].astype(np.float64)
dados['taxa_limpeza'] = dados['taxa_limpeza'].astype(np.float64)
```

# Removendo caracteres com regex

```python
#Removendo caracteres desnecessários com regex

dados['descricao_local'][3169]#selecionando a linha,
```

```python
#Fazendo substituicao com o metodo replace

dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']','', regex=True) #apenas para selecionar os valores [^a-zA-Z0-9\-\']
```

```python
#Atribuindo novamente a coluna

dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']','', regex=True)
```

```python
#Removendo hífens fora de palavra compostas

dados['descricao_local'].str.replace('(?<!\w)-(?!\w)','', regex=True)
```

```python
#Atribuindo a coluna

dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)','', regex=True)
```

Criando tokens, tokenização de strings

```python
#Tokenizando strings

dados['descricao_local'].str.split()
```

```python
#Atribuindo

dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
dados.head()
```

```python
#removendo caracteres que nao ajudam na tokenização

dados['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
#Atribuindo modificação

dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
#concluindo tokenização

dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

```python
# Transformamos o texto em letras minúsculas
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
# Substituimos os caractere especiais
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
# Transformamos o texto em lista, formando o token
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados.head()
```

# Transformando dados para tempo, trabalhando com dados datetime

```python
#transformado base de dados em dataframe

dt_data = pd.read_json('/content/moveis_disponiveis.json')
dt_data.head()
```

```python
dt_data.info()
```

```python
#aplicando o metodo to.datetime (alterando para datetime no formato/tipo de arquivo)

pd.to_datetime(dt_data['data'])
```

```python
#Atribuindo

dt_data['data'] = pd.to_datetime(dt_data['data'])
```

```python
dt_data.info()
```

```python
#Manipulando dados temporais
#verificando os dados
dt_data.head()
```

```python
#focando na coluna vaga diponível

dt_data['data'].dt.strftime('%Y-%m') #selecionando apenas o ano e o mês
```

```python
#agrupando os valores através do group by informando as vagas disponiveis por mes

dt_data.groupby(dt_data['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
```

```python
#deixando os dados como subset

subset = dt_data.groupby(dt_data['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
subset
```
